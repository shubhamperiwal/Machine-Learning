```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Point and Line Plots

```python
days = list(range(0,22,3))
temp_min = [23, 25, 21, 27, 30, 36, 26, 30]
temp_max = [24, 30, 23, 29, 34, 38, 27, 31]
```

```python
plt.plot(days, temp_min,
        days, temp_min, 'ob',
        days, temp_max,
        days, temp_max, 'or')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

### Spines

```python
plt.plot(days, temp_min,
        days, temp_min, 'ob',
        days, temp_max,
        days, temp_max, 'or')
plt.xlabel("Days")
plt.ylabel("Temperature")

#Used to manipulate spines and ticks
ax = plt.gca()
#Remove right and top spine
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```

```python
plt.plot(days, temp_min,
        days, temp_min, 'ob',
        days, temp_max,
        days, temp_max, 'or')
plt.xlabel("Days")
plt.ylabel("Temperature")

#Used to manipulate spines and ticks
ax = plt.gca()
#Remove right and top spine
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_bounds(22.5, 37.5)
plt.show()
```

### Ticks 

```python
plt.plot(days, temp_min,
        days, temp_min, 'ob',
        days, temp_max,
        days, temp_max, 'or')
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.xticks(range(0,25,2))
plt.show()
```

```python
plt.plot(days, temp_min,
        days, temp_min, 'ob',
        days, temp_max,
        days, temp_max, 'or')
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.xticks(range(0,28,7),['W1', 'W2', 'W3', 'W4'])
plt.show()
```

### Legends

```python
plt.plot(days, temp_min)
plt.plot(days, temp_min, 'ob', label='min')
plt.plot(days, temp_max)
plt.plot(days, temp_max, 'or', label='max')
plt.legend(loc='best')
plt.xlabel("Days")
plt.ylabel("Temperature")

ax = plt.gca()

plt.show()
```

### Annotation

```python
plt.plot(days, temp_min,
        days, temp_min, 'ob',
        days, temp_max,
        days, temp_max, 'or')
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.annotate('Temperature difference\nacross month', xy=(2, 35))

plt.show()
```

### Subplots

```python
fig = plt.figure(figsize=(8,6))
sub1 = plt.subplot(2,2,1)
sub2 = plt.subplot(2,2,2)
sub3 = plt.subplot(2,2,3)
sub4 = plt.subplot(2,2,4)

sub1.plot(days, temp_max, 'or')
sub1.set_xticks(()) #Remove XTicks
sub1.set_yticks(()) #Remove XTicks
sub1.set_title('temp max point plot')

sub2.plot(days, temp_max, 'r')
sub2.set_xticks(()) #Remove XTicks
sub2.set_yticks(()) #Remove XTicks
sub2.set_title('temp max line plot')

sub3.plot(days, temp_min, 'ob')
sub3.set_xticks(()) #Remove XTicks
sub3.set_yticks(()) #Remove XTicks
sub3.set_title('temp min point plot')

sub4.plot(days, temp_min, 'r')
sub4.set_xticks(()) #Remove XTicks
sub4.set_yticks(()) #Remove XTicks
sub4.set_title('temp min line plot')
```

### Histograms

```python
gaus_numbers = np.random.normal(size=1000)
```

```python
#Bins: For number of bins
#Edge color: For colours between bins
#Normed: Normalise values so sum is 1
plt.hist(gaus_numbers, bins=10, color='r',
        edgecolor='black', normed=True)
```

```python
#Cumulative: Self-explanatory
plt.hist(gaus_numbers, bins=10, color='r',
        edgecolor='black', normed=True, 
        cumulative=True)
```

### Contour Plots

```python
#Create data for contour plot
xlist = np.linspace(-1,1,3)
ylist = np.linspace(-1,1,3)
print("XList: ",xlist)

X, Y = np.meshgrid(xlist, ylist)
print("\nX: ", X, type(X)) #X and Y are numpy arrays

#Get distance between the 2 points
Z = np.sqrt(X**2 + Y**2)
print("\nZ: ",Z)
```

```python
plt.contour(X,Y,Z)
```

```python
cp = plt.contour(X,Y,Z)
plt.clabel(cp, inline=True, fontsize=10)
```

```python
cp = plt.contour(X,Y,Z)
#Add the labels
plt.clabel(cp, inline=True, fontsize=10)
#Colour fill
plt.contourf(X,Y,Z)
```

```python
level = [0.2, 0.6, 0.8, 1.7, 1.9, 2.9, 3.7]
cp = plt.contour(X,Y,Z,level, colors='k')
plt.clabel(cp, inline=True, fontsize=10)
plt.contourf(X,Y,Z)
```
